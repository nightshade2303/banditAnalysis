```python
def compute_density(x_data, y_data, bins):
    x_flat = x_data.flatten()
    y_flat = y_data.flatten()
    
    # Remove NaN values
    valid_indices = ~np.isnan(x_flat) & ~np.isnan(y_flat)
    x_valid = x_flat[valid_indices]
    y_valid = y_flat[valid_indices]
    
    # Compute 2D histogram
    hist, x_edges, y_edges = np.histogram2d(x_valid, y_valid, bins=bins, density=True)
    return hist.T, x_edges, y_edges

def plot_density_heatmap(x_positions, y_positions, ax, bins=50):
    """
    Plot probability density heatmap of x and y positions on a given axis.
    
    Parameters:
    - x_positions: List of 2D arrays containing x positions (each array can have NaN values).
    - y_positions: List of 2D arrays containing y positions (corresponding arrays to x_positions).
    - ax: Matplotlib axis object to plot on.
    - bins: Number of bins to use for the 2D histogram (default is 50).
    """
    # Function to calculate density ignoring NaNs

    
    # Calculate density for each dataset
    density_maps = []
    for x_data, y_data in zip(x_positions, y_positions):
        density, x_edges, y_edges = compute_density(x_data, y_data, bins)
        density_maps.append(density)
    
    # Sum all density maps to get the combined density
    combined_density = np.sum(density_maps, axis=0)
    
    # Plotting the heatmap on the provided axis
    im = ax.imshow(combined_density, extent=[x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]],
                   origin='lower', cmap='hot')
    ax.set_xlabel('X Position')
    ax.set_ylabel('Y Position')
    
    # Add colorbar
    cbar = ax.figure.colorbar(im, ax=ax, label='Density')
    
    return im, cbar


f2, ax2 = plt.subplots(1,2, figsize=[12,6])

plot_density_heatmap(shelter_xpos, shelter_ypos, ax2[1])
plot_density_heatmap(noshelter_xpos, noshelter_ypos, ax2[0])
```

```python
frame_times = annotated[(annotated['Behavior type']=='STOP') & (annotated['Behavior']=='Bout')
          & (annotated['Subject']=='neo_day2') & (annotated['cond']=='noshelter')]['Time']

import cv2
import numpy as np

def get_frames_around_times(video_path, frame_times, buffer_seconds, fps):
    cap = cv2.VideoCapture(video_path)
    segments = []
    
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    buffer_frames = int(buffer_seconds * fps)
    
    for frame_time in frame_times:
        start_frame = max(0, int((frame_time - buffer_seconds) * fps))
        end_frame = min(total_frames, int((frame_time + buffer_seconds) * fps))
        cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
        
        frames = []
        for _ in range(end_frame - start_frame):
            ret, frame = cap.read()
            if not ret:
                break
            frames.append(frame)
        segments.append(frames)
    
    cap.release()
    return segments, video_fps

def write_video(frames, output_path, fps):
    height, width, _ = frames[0].shape
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), fps, (width, height))
    for frame in frames:
        out.write(frame)
    out.release()

def create_video_grid(segments, output_path, grid_size, fps):
    frame_count = min(len(segment) for segment in segments)
    
    height, width, _ = segments[0][0].shape
    grid_height, grid_width = grid_size
    
    frame_height = grid_height * height
    frame_width = grid_width * width
    
    output_frames = []
    
    for i in range(frame_count):
        grid_frame = np.zeros((frame_height, frame_width, 3), dtype=np.uint8)
        for row in range(grid_height):
            for col in range(grid_width):
                segment_index = row * grid_width + col
                if segment_index < len(segments):
                    video_frame = segments[segment_index][i]
                    grid_frame[row*height:(row+1)*height, col*width:(col+1)*width, :] = video_frame
        output_frames.append(grid_frame)
    
    write_video(output_frames, output_path, fps)


# Example usage:
video_path = r'D:\final project\neo_day2\noshelter\Neo_noshelter_day_topView2024-06-18T12_31_58.avi'
buffer_seconds = 5  # 5 seconds before and after
grid_size = (2, 3)  # 2x2 grid

segments, fps = get_frames_around_times(video_path, frame_times, buffer_seconds, fps=30)
output_path = 'output_grid_noshelter.mp4'
create_video_grid(segments, output_path, grid_size, fps)
```

```python
PETH_l = []
for _vars in TENSS_generator(path):
    (data_path, sorting, centroid_df, sideView_ROI_df, topView_ROI_df, sideViewTimestamps_df, topViewTimestamps_df,
     boris_df, topView, sideView, unit_df, good_ids, st_d, recording, falling, unit, st) = _vars
    
    start_inds = boris_df.loc[(boris_df['Behavior'] == 'Bout') & (boris_df['Behavior type'] == 'START'), 'Image index'].values
    stop_inds = boris_df.loc[(boris_df['Behavior'] == 'Bout') & (boris_df['Behavior type'] == 'STOP'), 'Image index'].values
    
    for start_i, stop_i in zip(start_inds, stop_inds): # in frames
        start_sample = falling[start_i]
        start_PETH_l_pre = len(st[(st > (start_sample - 1 * 30_000)) & (st <= start_sample)]) # Hz
        start_PETH_l_post = len(st[(st > start_sample) & (st <= (start_sample + 1 * 30_000))]) # Hz
        stop_sample = falling[stop_i]
        stop_PETH_l_pre = len(st[(st > (stop_sample-1*30_000)) & (st<=stop_sample)]) # Hz
        stop_PETH_l_post = len(st[(st>stop_sample)&(st<=(stop_sample+1*30_000))]) # Hz

        PETH_l.append(pd.DataFrame({'unit': unit,
            "condition": os.path.basename(data_path),
            'FR': start_PETH_l_pre,
            "timepoint": "pre start aligned"}, index=[0]))
        
        PETH_l.append(pd.DataFrame({'unit': unit,
            "condition": os.path.basename(data_path),
            'FR': start_PETH_l_post,
            "timepoint":"post start aligned"}, index=[0]))
        
        PETH_l.append(pd.DataFrame({'unit': unit,
            "condition": os.path.basename(data_path),
            'FR': stop_PETH_l_pre,
            "timepoint": "pre stop aligned"}, index=[0]))
        
        PETH_l.append(pd.DataFrame({'unit': unit,
            "condition": os.path.basename(data_path),
            'FR': stop_PETH_l_post,
            "timepoint":"post stop aligned"}, index=[0]))

PETH_df = pd.concat(PETH_l)



sns.catplot(data=PETH_df, kind="strip", col="condition", x="timepoint", y="FR", col_wrap=2, height=3)
```
