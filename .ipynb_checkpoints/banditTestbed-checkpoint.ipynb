```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
sns.set_theme()
sns.set_style("ticks")
```

```python
def fxn(mean, arms):
    x = np.linspace(1, arms, arms)
    sig = 1.75/2
#     amp = 1/(sig*np.sqrt(2*np.pi))
    amp = 0.7
    vo = 0.1
    gx = (amp*np.exp(-0.5*((x-mean)**2)/(sig**2)))+vo
#     gx = np.random.permutation(gx)
    return gx
```

```python
# np.random.seed(4231)
n_arms = 4
arms = list(range(n_arms))

# print(prob_arms)
# rew_val = np.ones(n_arms)
# prob_arms = np.ones(n_arms)

# thompson sampling 
# update al, be
# al[action] = al[action] + reward
# be[action] = be[action] + 1 - reward
# for arm in range(len(arms)):
# theta[arm] = np.random.beta(al[arm],be[arm])
# 
# action = argmax (theta)

sess_mean = np.random.randint(n_arms)
gx = fxn(sess_mean+1, n_arms)
prob_arms = np.copy(np.random.permutation(gx))
print(prob_arms)
eps = 0.2
alpha = 0.2
gamma = 0.2
c = 0.2
shift_prob = 1

trials = 150
sessions = 10
temperature = 0.1
inv_temp = 1/temperature

value = np.zeros(n_arms)
prob_action = np.zeros(n_arms)
actions = []
nt = np.zeros(n_arms)
reward = 0
for i in range(trials):
    if len(actions) == 0:
        actions = [np.random.randint(n_arms)]
    av_actions = list(range(n_arms))
    if reward==1:
        action = actions[-1]
    else:
        shift = np.random.uniform(0,1)
        if shift<=shift_prob:
            av_actions.pop(actions[-1])
            action = np.random.choice(av_actions)
        else:
            action = actions[-1]
    actions.append(action)
    ran = np.random.uniform(0,1)
    if ran <= prob_arms[action]:
        reward = 1
    else: 
        reward = 0
    
    value[int(action)] = value[int(action)] + alpha * (reward - value[int(action)])   
```

```python
t1=[]
for ind, x in enumerate(actions):
    try:
        t1_port=actions[ind+1]
        t1.append(t1_port)            
    except IndexError as e:
        pass

# for all i, j, compute: 
# how many times does it go from i to j 

prob_array=np.zeros((n_arms,n_arms))

for p in range(len(t1)):
    for i in range(n_arms):
        for j in range(n_arms):
            temp = np.zeros((n_arms, n_arms))
            if ((actions[p]==i) & (t1[p]==j)):
                temp[i,j]=1
                prob_array[i,j]+= temp[i,j]

prob_array/=np.sum(prob_array)
g = sns.heatmap(prob_array, cmap = 'Blues')
from matplotlib.patches import Rectangle
g.add_patch(Rectangle((sess_mean, sess_mean), 1, 1, fill=False, edgecolor='r', lw=3))
g.invert_yaxis()
```

```python
# session variables
import numpy as np
arms = 4
trials = 200
inv_temp = 0.1
# alpha = np.array([0.1, 0.2, 0.3, 0.4])
alpha = np.full((arms, arms), 0.2)
np.fill_diagonal(alpha, 5)
alpha *= 0.1
l_ars = []

q = np.ones(arms)*(1/arms)
def fxn(mean, arms, permute = False):
    x = np.linspace(1, arms, arms)
    sig = 1.75/2
#     amp = 1/(sig*np.sqrt(2*np.pi))
    amp = 0.7
    vo = 0.1
    gx = (amp*np.exp(-0.5*((x-mean)**2)/(sig**2)))+vo
    if permute:
        gx = np.random.permutation(gx)
    return gx
    
rp = fxn(np.random.choice(arms), arms, True)

for t in range(trials):
    # remember prev_a
    a_prev = a
    
    # specify action
    p = [(np.exp(q[arm]*inv_temp)) / np.sum(np.exp(q*inv_temp)) for arm in range(arms)]
    actions = np.random.multinomial(1, p)
    a = actions.nonzero()[0][0]
    
    # get reward
    rand = np.random.uniform(0, 1)
    r = 1 if rand <= rp[a] else 0
    
    # update value of chosen arm
    delta = r - q[a]
    q[a] = q[a] + delta*alpha[a_prev, a]

    # devalue unchosen arms?
    # for an in range(arms):
    #     if an != a:
    #         q[an] = q[an] - delta*alpha[a_prev, an]
    
    # move on
    l_ars.append([a, r])
    print(a, r, q)
```
