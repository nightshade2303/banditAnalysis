```python
# WITH BOX #2 BECAUSE OF RECENT BIAS STUFF get t-1 th trial, get t - n th trial combinations
trialsinsess = 150
hist = 4
filtered = sessdf.groupby('session#').filter(lambda x: x.reward.size >= trialsinsess)

# prepping df - 4th port as invalid
# filtered['valid'] = np.zeros(len(filtered))
# filtered['valid'] = filtered['valid'].where(filtered['port'].isin([4.0]), 1)
# filtered = filtered[filtered['valid']==1.0]

filtered['choice_t0'] = filtered.port.values
for i in range(1,hist): 
    filtered['choice_t'+str(i)] = filtered.groupby('session#').port.shift(i)
    filtered['shift_t'+str(i-1)] = filtered['choice_t'+str(i)]==filtered['choice_t'+str(i-1)]
    filtered['shift_t'+str(i-1)] = filtered['shift_t'+str(i-1)].replace({True: 0, False: 1})
    filtered['reward_t'+str(i)] = filtered.groupby('session#').reward.shift(i)
#     filtered['noreward_t'+str(i)] = filtered['reward_t'+str(i)].replace({0:1, 1:0})

filtered = filtered.dropna()
filtered.groupby('session#').get_group(7).shape
```

```python
# get qval for each trial?
filtered['p_qvalue'] = np.zeros(filtered.shape[0])
filtered['qvalue'] = np.ones(filtered.shape[0])*0.33
filtered[['p_qvalue', 'qvalue']]=filtered[['p_qvalue', 'qvalue']].astype('object')
x0 = (0.05, 0.05)
arms = 3
# compute q val for fixed alpha, tau 
alpha, tau = x0
ll = 0

# p = np.zeros(filtered.shape[0])
q = np.ones(arms)/arms

for sessnum, group in filtered.groupby('session#'):
    q = np.ones(arms)/arms
    
    for ind, trial in group.iterrows():

        # softmax prob of choosing actions
        invtemp=1/tau
        P = np.exp(invtemp*q)
        P = P/ np.sum(P)

        # which action on this trial
        a = trial['port']
        index = int(a-1)

        # probability of each action on this trial
        filtered['p_qvalue'].loc[ind] = list(P)

        # rewarded?
        r = trial['reward']

        # compute q value
        q[index] = q[index] + alpha*(r - q[index])
        filtered['qvalue'].loc[ind] = list(q)
# ll += np.nansum(np.log(p))
# nll = -ll
# return nll
```

```python
# build a null model with the same switches as the animal does, but with randomly drawn choices
# slightly faster

# Filter sessions with three unique choices
# each port should appear atleast twice - count of port# 1, 2, 3 > 2
# tic()
# nulldf=pd.DataFrame()
# for sessnum, group in filtered.groupby('session#'):
#     accepted = group.apply(lambda x: group.port.loc[:x.name].value_counts()>3, axis=1)
#     accepted = accepted[(accepted[1]==True)&(accepted[2]==True)&(accepted[3]==True)]
#     nulldf = pd.concat([nulldf, accepted])
# toc()

# newdf = filtered.loc[nulldf.index,:]

# get only the shift!!!!
newdf = newdf[newdf['shift_t0']==1]
```

```python
tic()
ports = np.array([1,2,3,4])
# Create an empty DataFrame with predefined data types
dtype = {"animal": str, "task": str, "session": int, "trial": int, "from_port": 'category', "to_port": 'category',
         "choicecount_t0": int, "p_action": float, "rewcount_t0": int, "rr_t0": float,
         'random_port': 'category', "p_random":float, 'random_rr_t0': float}
switchdf = pd.DataFrame(columns=dtype.keys()).astype(dtype)

# Calculate tm matrix
tm = np.array(pd.crosstab(newdf.choice_t1,
                          newdf.choice_t0,
                          normalize='index'))
print(tm)

for sessnum, group in newdf.groupby('session#'):

    for ind, row in group.iterrows():
        from_port = int(row.choice_t1)
        to_port = int(row.choice_t0)
        
        choicecount_t0 = len(filtered[filtered['port']==to_port].groupby('session#').get_group(sessnum)['port'].loc[:ind])
        
        p_action_t0 = choicecount_t0/len(filtered.groupby('session#').get_group(sessnum)['reward'].loc[:ind])
        
        rewcount_t0 = sum(filtered[filtered['port']==to_port].groupby('session#').get_group(sessnum)['reward'].loc[:ind])
        
        rr_t0 = rewcount_t0/len(filtered[filtered['port']==to_port].groupby('session#').get_group(sessnum)['reward'].loc[:ind])
        
        i,j = np.where(ports!=row.choice_t1)[0]
        p = (tm[int(row.choice_t1)-1, i], tm[int(row.choice_t1)-1, j])
        random_port = np.random.choice(np.where(ports!=row.choice_t1)[0]+1, p=p)
        p_random_t0 = p[np.where((np.where(ports!=row.choice_t1)[0])==random_port-1)[0][0]]
        try:
            random_port_rewcount = sum(filtered[filtered['port']==random_port].groupby('session#').get_group(sessnum)['reward'].loc[:ind]) 
            random_port_rr = random_port_rewcount/len(filtered[filtered['port']==random_port].groupby('session#').get_group(sessnum)['reward'].loc[:ind])
        except ZeroDivisionError:
            random_port_rr = 0
        except KeyError:
            random_port_rewcount = 0

        # Append the row as a DataFrame to switchdf
        row_df = pd.DataFrame({'session': sessnum, 'trial': ind, 'from_port': from_port, 'to_port': to_port,
                               'choicecount_t0': choicecount_t0, 'p_action': p_action_t0,
                               'rewcount_t0': rewcount_t0, 'rr_t0': rr_t0,
                               'random_port': random_port, 'p_random':p_random_t0,
                               'random_rr_t0': random_port_rr},
                              index=[0])
        switchdf = pd.concat([switchdf, row_df], ignore_index=True)

# Convert 'from_port' and 'to_port' columns to category data type
switchdf['from_port'] = switchdf['from_port'].astype('category')
switchdf['to_port'] = switchdf['to_port'].astype('category')

# Print the resulting DataFrame
toc()
```

```python
switchdf['allq']= newdf.apply(lambda x: x.qvalue, axis =1).reset_index()[0]
switchdf['allp']= newdf.apply(lambda x: x.p_qvalue, axis =1).reset_index()[0]
switchdf['qvalue'] = switchdf.apply(lambda x: x.allq[int(x.to_port)-1], axis =1)
switchdf['p_qvalue'] = switchdf.apply(lambda x: x.allp[int(x.to_port)-1], axis =1)
switchdf['random_qvalue']=switchdf.apply(lambda x: x.allq[int(x.random_port)-1], axis =1)
switchdf['p_random_qvalue']=switchdf.apply(lambda x: x.allp[int(x.random_port)-1], axis =1)
```
