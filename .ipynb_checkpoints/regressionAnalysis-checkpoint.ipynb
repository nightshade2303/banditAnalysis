```python
# for implementation of Sabatini paper (mice exhibit stochastic and efficient decision making)

filtered['rc_t1'] = pd.from_dummies(filtered[['sr', 'snr', 'dr', 'dnr']]).values
filtered['rc_t2'] = np.zeros(len(filtered))
filtered.loc[(filtered['choice_t2']==filtered['choice_t3']) & (filtered['reward_t2']==1), 'rc_t2'] = 'sr'
filtered.loc[(filtered['choice_t2']==filtered['choice_t3']) & (filtered['reward_t2']==0), 'rc_t2'] = 'snr'
filtered.loc[(filtered['choice_t2']!=filtered['choice_t3']) & (filtered['reward_t2']==1), 'rc_t2'] = 'dr'
filtered.loc[(filtered['choice_t2']!=filtered['choice_t3']) & (filtered['reward_t2']==0), 'rc_t2'] = 'dnr'

filtered['rc_t3'] = np.zeros(len(filtered))
filtered.loc[(filtered['choice_t3']==filtered['choice_t4']) & (filtered['reward_t3']==1), 'rc_t3'] = 'sr'
filtered.loc[(filtered['choice_t3']==filtered['choice_t4']) & (filtered['reward_t3']==0), 'rc_t3'] = 'snr'
filtered.loc[(filtered['choice_t3']!=filtered['choice_t4']) & (filtered['reward_t3']==1), 'rc_t3'] = 'dr'
filtered.loc[(filtered['choice_t3']!=filtered['choice_t4']) & (filtered['reward_t3']==0), 'rc_t3'] = 'dnr'
filtered['rc_t1'] = filtered['rc_t1'].replace({'sr':'S', 'snr':'s', 'dnr':'d', 'dr':'D'})
filtered['rc_t2'] = filtered['rc_t2'].replace({'sr':'S', 'snr':'s', 'dnr':'d', 'dr':'D'})
filtered['rc_t3'] = filtered['rc_t3'].replace({'sr':'S', 'snr':'s', 'dnr':'d', 'dr':'D'})

filtered['hist_t3'] = np.zeros(len(filtered))
filtered.loc[:,['hist_t3']] = filtered['rc_t1']+filtered['rc_t2']+filtered['rc_t3']

filtered['rc_t4'] = np.zeros(len(filtered))
filtered.loc[(filtered['choice_t4']==filtered['choice_t5']) & (filtered['reward_t4']==1), 'rc_t4'] = 'S'
filtered.loc[(filtered['choice_t4']==filtered['choice_t5']) & (filtered['reward_t4']==0), 'rc_t4'] = 's'
filtered.loc[(filtered['choice_t4']!=filtered['choice_t5']) & (filtered['reward_t4']==1), 'rc_t4'] = 'D'
filtered.loc[(filtered['choice_t4']!=filtered['choice_t5']) & (filtered['reward_t4']==0), 'rc_t4'] = 'd'

filtered['hist_t4'] = np.zeros(len(filtered))
filtered.loc[:,['hist_t4']] = filtered['rc_t1']+filtered['rc_t2']+filtered['rc_t3']+filtered['rc_t4']

formula = 'shift_t0 ~ -1+hist_t3'

model = smf.glm(formula = formula, data = filtered, family = sm.families.Binomial()).fit()

reg_df_t3 = pd.DataFrame()
reg_df_t3['name'] = model.model.exog_names
reg_df_t3['coef'] = np.array(model.params)
reg_df_t3['err'] = np.array(model.bse)
reg_df_t3_sorted = reg_df_t3.sort_values('coef')

fig = plt.figure(figsize=(21,10))
ax = plt.subplot(111)

ax.bar(x = 'name', height = 'coef', data = reg_df_t3_sorted, yerr='err', color = 'grey')
sns.despine()

ax.set_xticklabels([i.split('[')[-1].split(']')[0] for i in reg_df_t3_sorted['name']], rotation=90)
ax.set_ylabel('Coeff(Different)')
plt.text(x = 27.5, y = 1, s = f'Model $R^2$ = {round(model.pseudo_rsquared(), 3)}', fontsize = 'x-large')
plt.title('GLM for choice and reward history of 3 trials', fontsize = 16)
plt.tight_layout()
# plt.savefig('plots/25082023/glm_t3hist.png', dpi = 600)
```

```python
# implementation of kevin miller paper (from predictive models to cognitive models)
# choice history - +1 same, -1 different
# reward history - +1 rewarded, -1 unrewarded
# choicexreward - +1 or -1 depending on combination
km_p = pd.DataFrame()
i=0
km_p['shift_t'+str(i)] = filtered['shift_t'+str(i)].replace({'s':0, 'd':1}).astype(int)
for i in range(1,20):
    km_p['shift_t'+str(i)] = filtered['shift_t'+str(i)].replace({'s':-1, 'd':1}).astype(int)
    km_p['reward_t'+str(i)] = filtered['reward_t'+str(i)].replace({0:-1}).astype(int)
    km_p['shxrew_t'+str(i)] = km_p[('shift_t'+str(i))]*km_p[('reward_t'+str(i))]

formula_str = ['shift_t'+str(i)+'+reward_t'+str(i)+'+shxrew_t'+str(i)+'+'for i in range(1,20)]
formula_str = str(''.join(formula_str))[:-1] 
formula = f'shift_t0 ~ {formula_str}'

model = smf.glm(formula = formula, data = km_p, family = sm.families.Binomial()).fit()

reg_df_t5 = pd.DataFrame()
reg_df_t5['name'] = model.model.exog_names
reg_df_t5['coef'] = np.array(model.params)
reg_df_t5['err'] = np.array(model.bse)

fig = plt.figure(figsize=(10,5))
ax = plt.subplot(111)
ax.bar(model.model.exog_names, model.params, yerr = model.bse)
sns.despine()
ax.set_xticklabels(model.model.exog_names, rotation= 90)
ax.set_ylabel('Coefficient')
```

```python
# multiclass regression 
# only choices, only first 100 trials
# filtered = sessdf.groupby('session#').filter(lambda x: x.reward.size >= trialsinsess)
# for sessnum, group in filtered.groupby('session#'):
#      group = group.iloc[:100, :]
bic_c = []
prsq_c =[]
formula = 'choice_t0 ~ choice_t1'
model = smf.mnlogit(formula = formula,
                    data = filtered[filtered.shift_t0==1]).fit()
bic_c.append(model.bic)
prsq_c.append(model.prsquared)
for i in range(2,20):
    formula+='+choice_t'+str(i)+'+'
    formula = formula[:-1]
    model = smf.mnlogit(formula = formula, data = filtered[filtered.shift_t0==1]).fit()
#     print(model.summary(), model.aic, model.bic)
    bic_c.append(model.bic)
    prsq_c.append(model.prsquared)
```
