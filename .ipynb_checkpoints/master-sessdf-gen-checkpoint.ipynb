```python
def find_all_trial_pokes(df, start_event=81, event_list=[18, 19, 20, 21, 20, 37, 17]):
    
    # Reset the index to ensure it's continuous integers
    df = df.reset_index(drop=True)
    
    # Find the row where event == 81
    start_rows = df[df['event'] == start_event].index

    if len(start_rows) == 0:
        return None, None  # No start event found
    
    results = [] # to store start and end pairs

    for start_row in start_rows:
        # Move upward to find the first row with event in the range [18, 21] and value == 1
        for i in range(start_row, -1, -1):
            if df.loc[i, 'event'] in event_list and df.loc[i, 'value'] == 1:
                first_event_row = i
                break
        else:
            continue  # If no event in range found, continue to next '81'

        # Move downward to find the first row with the same event but value == 0
        for j in range(first_event_row, len(df)):
            if df.loc[j, 'event'] == df.loc[first_event_row, 'event'] and df.loc[j, 'value'] == 0:
                # Append the start and end rows from the original DataFrame
                results.append(df.loc[first_event_row])
                results.append(df.loc[j])
                break

    # Create a DataFrame from the results
    result_df = pd.DataFrame(results)

    return result_df
```

```python
import pandas as pd
import numpy as np

def find_all_trial_pokes_v2(df, start_event=81, event_list=[18, 19, 20, 21]):
    # Reset the index to ensure continuity
    df = df.reset_index(drop=True)

    # Extract relevant columns as NumPy arrays for faster operations
    event_array = df['event'].to_numpy()
    value_array = df['value'].to_numpy()

    # Identify indices of start_event and event_list
    start_indices = np.where(event_array == start_event)[0]
    relevant_indices = np.where(np.isin(event_array, event_list))[0]

    if len(start_indices) == 0 or len(relevant_indices) == 0:
        return None  # No start events or relevant events found

    # Initialize lists to store results
    start_rows = []
    end_rows = []

    for start_index in start_indices:
        # Find the largest relevant index less than the start index
        earlier_relevant_indices = relevant_indices[relevant_indices < start_index]
        if earlier_relevant_indices.size == 0:
            continue
        first_event_index = earlier_relevant_indices[-1]

        # Ensure the value at the first event is 1
        if value_array[first_event_index] != 1:
            continue

        # Find the first matching event with value == 0 after the first_event_index
        later_indices = np.where(
            (event_array[first_event_index + 1:] == event_array[first_event_index]) & 
            (value_array[first_event_index + 1:] == 0)
        )[0]
        if later_indices.size == 0:
            continue
        end_index = first_event_index + 1 + later_indices[0]

        # Append start and end indices
        start_rows.append(first_event_index)
        end_rows.append(end_index)

    # Combine start and end indices
    all_indices = np.array(start_rows + end_rows)

    # Extract rows in bulk using NumPy indexing
    result_df = df.iloc[all_indices].reset_index(drop=True)

    return result_df
```

```python
def gen_camera_df(event_files, sessdf):
    
    # generate dataframe of data from camera event files
    data = []
    cam_frame_start = []
    
    for filename in event_files:
        events = pd.read_csv(filename, names=['pinstatus', 'camtime', 'pitime'], skiprows=1)
        with open(filename.rstrip('events')+'frames') as f:
            cam_frame_start.append(f.readline().strip('\n'))
        frames = pd.read_csv(filename.rstrip('events')+'frames', skiprows=1)
        events['filename'] = os.path.basename(filename)
        data.append(events)

    cam_frame_start = [float(x) for x in cam_frame_start]
    
    # generate dataframe of data from camera frames
    camera_df = pd.concat(data, ignore_index=True)
    # frames = pd.concat(frame_data)
    camera_df['rec_time'] = camera_df.filename.apply(lambda x: x[4:9]).astype(int)
    camera_df['human_time'] = pd.to_datetime(camera_df.pitime, unit='s')
    camera_df['vid_start'] = np.nan
    camera_df.iloc[camera_df.groupby('filename').head(1).index, -1] = cam_frame_start
    camera_df['vid_start'].ffill(inplace = True)
    sessdf['human_time'] = np.array(sessdf['datetime'], dtype='datetime64[ns]')
    
    return camera_df
```

```python
import pandas as pd

def find_matches(camtime: pd.DataFrame, sample_date_diff: pd.DataFrame) -> pd.DataFrame:
    # Initialize list to store matches
    matches = []
    
    # Keep track of the indices in sample_date_diff that have been used
    used_indices = set()

    # Iterate over the camtime DataFrame
    for camtime_index, camtime_value in camtime.itertuples():
        # Find matching rows in smtime within Â±1 range, excluding previously used rows
        matching_rows = sample_date_diff[(sample_date_diff['smtime'] >= camtime_value - 1) & 
                                         (sample_date_diff['smtime'] <= camtime_value + 1) &
                                         (~sample_date_diff.index.isin(used_indices))]
        
        if not matching_rows.empty:
            # Store the first available match based on order
            match = matching_rows.iloc[0]
            
            # Record the match and mark this smtime index as used
            matches.append((camtime_index, camtime_value, match.name, match['smtime']))
            used_indices.add(match.name)

    # Convert the matches to a DataFrame
    matches_df = pd.DataFrame(matches, columns=['camtime_index', 'camtime_value', 'smtime_index', 'smtime_value'])

    return matches_df
```

```python
import pandas as pd

def find_matches_v2(camtime: pd.DataFrame, sample_date_diff: pd.DataFrame) -> pd.DataFrame:
    # Sort both DataFrames by relevant columns
    camtime = camtime.sort_values(by='camtime').reset_index()
    sample_date_diff = sample_date_diff.sort_values(by='smtime').reset_index()

    # Initialize pointers and matches list
    matches = []
    smtime_idx = 0  # Pointer for sample_date_diff

    # Iterate over camtime
    for cam_idx, cam_value in camtime[['index', 'camtime']].itertuples(index=False):
        while smtime_idx < len(sample_date_diff) and sample_date_diff.loc[smtime_idx, 'smtime'] < cam_value - 1:
            smtime_idx += 1  # Move pointer forward if smtime is too small

        # Check if current smtime is within range
        if smtime_idx < len(sample_date_diff) and cam_value - 1 <= sample_date_diff.loc[smtime_idx, 'smtime'] <= cam_value + 1:
            matches.append((cam_idx, cam_value, sample_date_diff.loc[smtime_idx, 'index'], sample_date_diff.loc[smtime_idx, 'smtime']))
            smtime_idx += 1  # Advance pointer to avoid reusing the same smtime

    # Convert matches to DataFrame
    matches_df = pd.DataFrame(matches, columns=['camtime_index', 'camtime_value', 'smtime_index', 'smtime_value'])

    return matches_df
```

#### filter camera_df based on matches

```python
import pandas as pd

def filter_camera_df(matches_df: pd.DataFrame, camera_df: pd.DataFrame) -> pd.DataFrame:
    """
    This function filters the camera_df based on the indices found in the matches_df
    and includes the previous index for each matched row, if applicable.

    :param matches_df: DataFrame containing 'camtime_index' column with matched indices
    :param camera_df: DataFrame to be filtered
    :return: Filtered DataFrame containing the matched rows and the row before each matched index
    """
    # Extract indices from the matches_df DataFrame
    indices = matches_df['camtime_index'].tolist()

    # Create a set of indices including each index and the one before it
    indices_with_previous = set(indices)  # Use a set to handle unique values

    # Add the index before each index to the set
    for idx in indices:
        if (idx - 1) in camera_df.index:  # Ensure the previous index is within the range of camera_df
            indices_with_previous.add(idx - 1)

    # Convert the set to a sorted list of indices
    sorted_indices = sorted(indices_with_previous)

    # Filter the camera_df based on the sorted indices
    filtered_camera_df = camera_df.loc[sorted_indices]

    return filtered_camera_df
```
