```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
dataset1 = pd.read_excel('C:/Users/dlab/Downloads/lily080424dcz2.xlsx')
```

```python
dataset1['Behavior_t1'] = dataset1.Behavior.shift(-1)
dataset1
```

```python
%matplotlib qt
fig = plt.figure(figsize = (7,6.5))
order = ['Advance', 'Pronate', 'Grasp', 'Pinch Grasp', 'Retract', 'Supinate', 'Fling pasta in air', 'Pasta enters box','Transfer to mouth', 'Head turn']
c = pd.crosstab(dataset1.Behavior, dataset1.Behavior_t1, normalize = 'index')
c = c[order]
c = c.T[order]
sns.heatmap(c.T, cmap = 'BuPu', annot = True, square = True,
            fmt = '.1f')
plt.xlabel('Behavior state at t+1')
plt.ylabel('Behavior state at t')
plt.tight_layout()
```
