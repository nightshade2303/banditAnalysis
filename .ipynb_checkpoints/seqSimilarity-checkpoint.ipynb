```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import enchant
```

```python
ports = np.random.choice(['a','b','c','d'], size = (4,40))

# generate all possible substrings
def substrings(s):
# s = ports[0]
    count = 0
    ss = list(np.zeros(int(len(s)*(len(s)+1)/2), dtype = str))
    # start index
    for i in range(len(s)):
        subStr = ""

        # generate substring
        for j in range(i,len(s)):
            subStr += s[j]
            ss[count] = str(subStr)
            count+=1
    return np.unique(ss)


for i in range(ports.shape[0]):
    ss = substrings(ports[i])
    #np.apply_along_axis(substrings, axis=1, arr=ports), axis = 1)
#     print(ss.shape)
#     ss = np.hstack((temp,ss))
```

```python
ss1 = substrings(ports[0])
ss2 = substrings(ports[1])
```

```python
ss1 = list(ss1)+list(np.zeros(abs(ss1.shape[0]-ss2.shape[0]), dtype = str))
```

```python
np.array(ss1).shape, np.array(ss2).shape 
```

```python
unique_combinations = []
 
for i in range(len(ss1)):
    for j in range(len(ss2)):
        unique_combinations.append((ss1[i], ss2[j]))
```

```python
dist = []
for (i,j) in unique_combinations:
    dist.append(enchant.utils.levenshtein(i, j))
```

```python
sns.histplot(dist, discrete = True)
```

```python
df = pd.DataFrame(unique_combinations, columns = ['i', 'j'])
```
