```python
################# PROBABILITY FUNCTIONS ################
def fxn(mean, arms, sig):
    x = np.linspace(1, arms, arms)
#     sig = 1.75
#     amp = 1/(sig*np.sqrt(2*np.pi))
    amp = 0.700001
    vo = 0.10
    gx = (amp*np.exp(-0.5*((x-mean)**2)/(sig**2)))+vo
#     gx = np.random.permutation(gx)
    return gx

def cauchy(median, arms, scale, s):
#     cauchy = @(x, s, t)(1./(s*pi*(1+(((x-t)./s).^2))))
    x = np.linspace(1, arms, arms)
#     s = 1.5/2
    cauchy = ((1/(s*np.pi*(1+(((x-median)/s)**2))))*scale)+0.1;
    return cauchy
# a = list(map(int, df[df[0]==83][0:8][4]/10))
# plt.plot(a, 'o-')

sig = 1.75
scale = 3.5
s = 1.5

fig, ax1 = plt.subplots(4, 5)
fig.set_figheight(12)
fig.set_figwidth(17)
arms = 4
for mean in range(1,5):
    ax1[0,mean].plot(fxn(mean, arms, sig/2)*10,'o-')
    ax1[0,mean].plot(list(map(round,fxn(mean, arms, sig/2)*10)), 'o-')
    ax1[0,mean].plot(list(map(int,fxn(mean, arms, sig/2)*10)), 'o-')
    ax1[0,mean].set_title('Gaussian')
    ax1[1,mean].plot(cauchy(mean, arms, scale/2, s/2)*10, 'o-')
    ax1[1,mean].plot(list(map(round,cauchy(mean, arms, scale/2, s/2)*10)), 'o-')
    ax1[1,mean].plot(list(map(int,cauchy(mean, arms, scale/2, s/2)*10)), 'o-')
    ax1[1,mean].set_title('Cauchy')
arms = 8
for mean in range(1,5):
    ax1[2,mean].plot(fxn(mean, arms, sig)*10,'o-')
    ax1[2,mean].plot(list(map(round,fxn(mean, arms, sig)*10)), 'o-')
    ax1[2,mean].plot(list(map(int,fxn(mean, arms, sig)*10)), 'o-')
    ax1[2,mean].set_title('Gaussian')
    ax1[3,mean].plot(cauchy(mean, arms, scale, s)*10, 'o-')
    ax1[3,mean].plot(list(map(round,cauchy(mean, arms, scale, s)*10)), 'o-')
    ax1[3,mean].plot(list(map(int,cauchy(mean, arms, scale, s)*10)), 'o-')
    ax1[3,mean].set_title('Cauchy')
plt.legend(['value', 'rounded', 'scale down'])
plt.tight_layout()
```
